```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1 f(x) = -4x**2 + 17x - 14
a = -4
b = 17
c = -14

x = np.linspace(-5, 9.25)
y = -4 * x**2 + 17 * x - 14

plt.figure(figsize=(6, 6))
plt.plot(x, y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x) = -4x^2 + 17x - 14')
plt.axhline()
plt.axvline()
plt.show()
```

```python
#2 6sin^2(x) + 7cos(x) - 7 = 0

x = np.linspace(-5, 5)
y = 6 * np.sin(x)**2 + 7 * np.cos(x) - 7 

plt.figure(figsize=(6, 6))
plt.plot(x, y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции 6sin^2(x) + 7cos(x) - 7')
plt.axhline()
plt.axvline()
plt.show()
```
